## Función para obtener respuestas generadas por ChatGPT utilizando la API de OpenAI.
Este código define una función llamada "get_completion" que utiliza la API de OpenAI para obtener una respuesta generada por el modelo de lenguaje ChatGPT. El parámetro "prompt" se utiliza como entrada de usuario para iniciar la conversación.

La función crea una lista de mensajes donde el rol es "usuario" y el contenido es el prompt proporcionado. Luego, se realiza una llamada a la API de OpenAI para generar una respuesta utilizando el modelo especificado (por defecto, "gpt-3.5-turbo"). El parámetro "temperature" controla la aleatoriedad de la generación de respuestas.

En este caso, el valor del parámetro "temperature" es 0, lo que significa que se selecciona la opción más probable y determinista. Si se incrementa el valor, como por ejemplo a 0.8, las respuestas pueden ser más diversas y creativas, ya que se introduce más aleatoriedad en el proceso de generación.

La función devuelve el contenido de la primera opción de respuesta obtenida de la API de OpenAI.

Cabe destacar que se requiere una clave de API válida para utilizar la API de OpenAI y que el código presentado no incluye la configuración o importación de las bibliotecas necesarias, por lo que se asume que se han realizado previamente.

````python
openai.api_key = "sk-TU key"
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0,
    )
    return response.choices[0].message["content"]
````

## API Spotify
Este código utiliza la biblioteca "spotipy" para acceder a la API de Spotify y obtener enlaces de reproducción de canciones. Se define una función llamada "obtener_enlace_cancion" que busca una canción por nombre y artista en Spotify y devuelve su enlace de reproducción. Luego, se muestra una función llamada "Spotify" que recibe texto como entrada, lo procesa y muestra los enlaces de reproducción de las canciones encontradas.

````python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Credenciales de autenticación de la API de Spotify
client_id = 'TU ID'
client_secret = 'ID'

# Inicializar el cliente de autenticación
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spotify = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

# Función para obtener el enlace de reproducción de una canción
def obtener_enlace_cancion(nombre_cancion, nombre_artista):
    # Realizar una búsqueda de la canción utilizando el nombre y el artista
    resultados = spotify.search(q=f'track:{nombre_cancion} artist:{nombre_artista}', type='track', limit=1)

    # Obtener el enlace de reproducción de la primera canción encontrada
    if len(resultados['tracks']['items']) > 0:
        enlace_cancion = resultados['tracks']['items'][0]['external_urls']['spotify']
        return enlace_cancion

    return None

##esta funcion recibe el prompt para genenrar los links
def Spotify(texto):
  lineas = texto.split('\n')
  datos = [linea.split(', ') for linea in lineas]
  dataframe = pd.DataFrame(datos, columns=['nombre_cancion', 'nombre_artista'])
  for i in range(len(dataframe)):
    enlace_cancion = obtener_enlace_cancion(dataframe.iloc[i]['nombre_cancion'], dataframe.iloc[i]['nombre_artista'])
    print("Cancion ",i+1," :",enlace_cancion)



````
